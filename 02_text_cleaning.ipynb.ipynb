```python
import pandas as pd
import os
from pathlib import Path
import re

# Paths
txt_folder = "../data/philosophy_texts/txt/"
keyword_file = "../data/keywords.csv"
output_file = "../data/philosophy_keywords_extracted.csv"

# Load keywords
keywords_df = pd.read_csv(keyword_file)
keywords = keywords_df['keyword'].str.lower().tolist()

# Get all text files
txt_files = [f for f in os.listdir(txt_folder) if f.endswith(".txt")]

data = []

def clean_text(text):
    # Basic cleaning: remove weird breaks, extra spaces, etc.
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower()

# Extract relevant paragraphs
for txt_file in txt_files:
    author = txt_file.replace(".txt", "").replace("_", " ")
    print(f"Processing {author}...")
    
    with open(os.path.join(txt_folder, txt_file), "r", encoding="utf-8") as f:
        raw_text = clean_text(f.read())

    # Split into paragraphs
    paragraphs = re.split(r'\n{2,}', raw_text)
    
    for para in paragraphs:
        for kw in keywords:
            if f" {kw} " in f" {para} ":
                data.append({
                    "author": author,
                    "keyword": kw,
                    "paragraph": para.strip()
                })

# Save to CSV
df = pd.DataFrame(data)
Path("../data/").mkdir(parents=True, exist_ok=True)
df.to_csv(output_file, index=False)
print(f"\n✅ Saved keyword-context data to {output_file}")
```
