```python
from pathlib import Path
import fitz
import os

# Corrected folders
pdf_folder = "../data/philosophy_pdfs/"
txt_folder = "../data/philosophy_texts/txt/"
Path(txt_folder).mkdir(parents=True, exist_ok=True)

# Book files
pdf_files = {
    "jung_psychological_types": "Psychological_Types_Carl_Jung.pdf",
    "russell_philosophy": "The_Problems_of_Philosophy_Bertrand_Russell.pdf",
    "nietzsche_zarathustra": "Thus_Spoke_Zarathustra_Friedrich_Nietzsche.pdf"
}

def extract_text_from_pdf(pdf_path):
    if not os.path.isfile(pdf_path):
        print(f"❌ File not found: {pdf_path}")
        return ""
    
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Process all PDFs
for name, filename in pdf_files.items():
    pdf_path = os.path.join(pdf_folder, filename)
    txt_path = os.path.join(txt_folder, f"{name}.txt")

    print(f"Extracting {filename}...")
    raw_text = extract_text_from_pdf(pdf_path)

    if raw_text:
        with open(txt_path, "w", encoding="utf-8") as f:
            f.write(raw_text)
        print(f"✅ Saved to {txt_path}\n")
```
