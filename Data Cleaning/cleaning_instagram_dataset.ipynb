```python
from pathlib import Path
import pandas as pd

# Set folder paths
raw_folder = Path(r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\pop_texts\instagram\instagram_comments")
cleaned_folder = Path(r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\pop_texts\instagram\instagram_comments_cleaned")
cleaned_folder.mkdir(parents=True, exist_ok=True)

# List all raw files (CSV and Excel)
raw_files = sorted([f for f in raw_folder.iterdir() if f.suffix in ['.csv', '.xlsx']])
cleaned_files = {f.stem.replace('_cleaned', '') for f in cleaned_folder.glob("*_cleaned.csv")}

# Get next unprocessed file
next_file = None
for file in raw_files:
    name = file.stem
    if name not in cleaned_files:
        next_file = file
        break

print(f"👉 Next file to clean: {next_file.name}" if next_file else "✅ All files cleaned.")
```

```python
# Load the selected file
input_path = next_file

if input_path.suffix == '.csv':
    df = pd.read_csv(input_path, dtype=str, encoding='utf-8', engine='python', on_bad_lines='skip')
elif input_path.suffix == '.xlsx':
    df = pd.read_excel(input_path, dtype=str, engine='openpyxl')
else:
    raise ValueError(f"🛑 Unsupported file type: {input_path.suffix}")

print(f"\n📂 Loaded: {input_path.name}")
print("🧾 Columns:", df.columns.tolist())
df.head()
```

```python
def convert_instagram_metric(value):
    """Convert Instagram metrics (1K -> 1000, 1M -> 1000000)"""
    if pd.isna(value):
        return 0
    value = str(value).replace(',', '')
    if 'K' in value:
        return int(float(value.replace('K', '')) * 1000
    elif 'M' in value:
        return int(float(value.replace('M', '')) * 1000000
    try:
        return int(float(value))
    except:
        return 0

# Convert all metric columns
metric_cols = [col for col in df.columns if any(x in col for x in ['views', 'likes', 'comments'])]
for col in metric_cols:
    df[col] = df[col].apply(convert_instagram_metric)

print("✅ Metrics converted to integers:")
display(df.head())
```

```python
import pandas as pd
from pathlib import Path

# Set folder paths
cleaned_folder = Path(r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\pop_texts\instagram\instagram_comments_cleaned")
output_path = cleaned_folder.parent / "instagram_comments_merged.csv"

# Get all cleaned files
cleaned_files = list(cleaned_folder.glob("*_cleaned.csv"))

if not cleaned_files:
    print("❌ No cleaned files found in the folder")
else:
    print(f"🔍 Found {len(cleaned_files)} cleaned files to merge")
    
    # Read and concatenate all files
    dfs = []
    for file in cleaned_files:
        try:
            df = pd.read_csv(file)
            dfs.append(df)
            print(f"✅ Loaded {file.name}")
        except Exception as e:
            print(f"❌ Failed to load {file.name}: {str(e)}")
    
    if dfs:
        merged_df = pd.concat(dfs, ignore_index=True)
        
        # Remove potential duplicates
        initial_count = len(merged_df)
        merged_df = merged_df.drop_duplicates(subset=['username', 'comment', 'comment_url'], keep='first')
        final_count = len(merged_df)
        
        # Save merged file
        merged_df.to_csv(output_path, index=False)
        print(f"\n🎉 Successfully merged {len(dfs)} files")
        print(f"📊 Initial rows: {initial_count:,}")
        print(f"📊 Final rows after deduplication: {final_count:,}")
        print(f"💾 Saved to: {output_path}")
    else:
        print("❌ No valid DataFrames to merge")
```
