Importing the right packages and dataset

```python
import pandas as pd

tiktok_path = r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\pop_texts\tiktok\tiktok_phil_data.csv"
df_tiktok = pd.read_csv(tiktok_path)

df_tiktok.head()
```

Clean and Rename column titles

```python
df_tiktok_cleaned = df_tiktok[[
    'uniqueId',
    'text',
    'diggCount',
    'createTimeISO',
    'videoWebUrl'
]].rename(columns={
    'uniqueId': 'username',
    'text': 'comment',
    'diggCount': 'like_count',
    'createTimeISO': 'time_posted',
    'videoWebUrl': 'comment_url'
})

df_tiktok_cleaned.head()
```

Clean the data types

Start with turning time_posted to datetime and like_count to interger

```python
df_tiktok_cleaned['time_posted'] = pd.to_datetime(df_tiktok_cleaned['time_posted'])
df_tiktok_cleaned.dtypes
```

Clean the comment text, first off by applying lowercasing

```python
# Save a copy of the original comment
df_tiktok_cleaned['comment_original'] = df_tiktok_cleaned['comment']

# Convert comment text to lowercase
df_tiktok_cleaned['comment'] = df_tiktok_cleaned['comment'].str.lower()

df_tiktok_cleaned[['comment_original', 'comment']].head()
```

cleaning for unnecessary emojis, urls or errors in the comment

```python
import re

def clean_text_keep_emojis(text):
    # Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    
    # Remove punctuation (except apostrophes)
    text = re.sub(r"[^\w\s'🧠💀😭😂🔥✨❤️😍🤔🤯😩😢😡😎🙏]", '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

df_tiktok_cleaned['comment'] = df_tiktok_cleaned['comment'].apply(clean_text_keep_emojis)

df_tiktok_cleaned[['comment_original', 'comment']].head()
```

checking and handling missing variables

```python
df_tiktok_cleaned.isnull().sum()
```

checking and removing the duplicates

```python
df_tiktok_cleaned.duplicated().sum()
```

```python
df_tiktok_cleaned['comment'].duplicated().sum()
```

From the above we see that there are no dublicate rows, only repeated comments from different people who feel the same way about a topic which makes sense in pop culture.

So we'll, keep them for now but will add a column to flag them and use both columns when it's time to analyze later

```python
df_tiktok_cleaned['is_duplicate_comment'] = df_tiktok_cleaned['comment'].duplicated()
df_tiktok_cleaned[['comment', 'is_duplicate_comment']].sample(5)
```

Save the new cleaned tiktok file

```python
df_tiktok_cleaned.to_csv("cleaned_tiktok_data.csv", index=False)
```
