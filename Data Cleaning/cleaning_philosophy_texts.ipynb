```python
import os
import pandas as pd
from datetime import datetime, timedelta
import random
import uuid

# Path to philosophy text files
input_dir = r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\philosophy_texts\txt"
output_dir = r"C:\Users\USER\Desktop\Projects\philosophy_and_popculture\data\philosophy_texts\txt"
os.makedirs(output_dir, exist_ok=True)

# Function to chunk raw text into paragraph-like quotes
def split_text_to_quotes(text, min_length=100, max_length=300):
    quotes = []
    current = ""
    for line in text.splitlines():
        line = line.strip()
        if not line:
            continue
        if len(current) + len(line) < max_length:
            current += " " + line
        else:
            if len(current) >= min_length:
                quotes.append(current.strip())
            current = line
    if len(current) >= min_length:
        quotes.append(current.strip())
    return quotes

# Metadata setup
def generate_row(quote, source_file):
    comment_original = quote.strip()
    comment = comment_original.lower()
    fake_username = f"@{uuid.uuid4().hex[:8]}"
    post_time = datetime.now() - timedelta(days=random.randint(0, 1000), hours=random.randint(0, 23))
    like_count = random.randint(0, 5000)
    platform = "text"
    source_type = "philosophy"
    return {
        "username": fake_username,
        "time_posted": post_time.isoformat(),
        "comment": comment,
        "comment_original": comment_original,
        "like_count": like_count,
        "comment_url": f"https://philosophy.fake/{source_file.replace('.txt', '')}/{fake_username}",
        "profile_url": f"https://philosophy.fake/user/{fake_username}",
        "platform": platform,
        "source_type": source_type
    }

# Process all text files
all_data = []
for filename in os.listdir(input_dir):
    if filename.endswith(".txt"):
        path = os.path.join(input_dir, filename)
        with open(path, "r", encoding="utf-8") as f:
            text = f.read()
            quotes = split_text_to_quotes(text)
            rows = [generate_row(q, filename) for q in quotes]
            all_data.extend(rows)

# Convert to DataFrame
df = pd.DataFrame(all_data)

# Save
output_path = os.path.join(output_dir, "philosophy_texts_merged.csv")
df.to_csv(output_path, index=False)

print(f"✅ Philosophy texts processed and saved: {df.shape}")
```
