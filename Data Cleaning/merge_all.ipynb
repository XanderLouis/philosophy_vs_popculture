```python
import pandas as pd

# ✅ Load datasets
df_phil = pd.read_csv("C:/Users/USER/Desktop/Projects/philosophy_and_popculture/data/philosophy_texts/philosophy_texts_merged.csv")
df_pop = pd.read_csv("C:/Users/USER/Desktop/Projects/philosophy_and_popculture/data/pop_texts/merged/popculture_texts_merged.csv", low_memory=False)

# ✅ Keep essential columns
df_phil_clean = df_phil[['comment', 'comment_original', 'platform', 'source_type']]
df_pop_clean = df_pop[['comment', 'comment_original', 'platform', 'source_type']]

# ✅ Merge both
df_combined = pd.concat([df_phil_clean, df_pop_clean], ignore_index=True)

# ✅ Save to corrected path
save_path = "C:/Users/USER/Desktop/Projects/philosophy_and_popculture/data/philosophy_vs_popculture.csv"
df_combined.to_csv(save_path, index=False)

print("✅ Combined dataset saved to:", save_path)
print("📊 Combined dataset shape:", df_combined.shape)
```
